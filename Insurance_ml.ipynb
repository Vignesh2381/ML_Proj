```python
from sklearn.model_selection import train_test_split
def splitt(x,y):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
    return x_train,y_train,x_test,y_test
```

```python
x=df.drop(labels=['charges'],axis=1)
y=df['charges']
y
```

```python
from sklearn.preprocessing import StandardScaler
def scalee(x_train,x_test):
    scale=StandardScaler()
    x_train=scale.fit_transform(x_train)
    x_test=scale.transform(x_test)
    return x_train,x_test
```

```python
x_train,x_test=scalee(x_train,x_test)
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.preprocessing import PolynomialFeatures
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.pipeline import make_pipeline

models={'LinearRegression':LinearRegression(),
       'Poly':make_pipeline(PolynomialFeatures(2),LinearRegression()),
       'Ridge':Ridge(alpha=1.0),
       'Lasso':Lasso(alpha=1.0),
       'DecisionTreeRegressor':DecisionTreeRegressor(),
       'RandomForestRegressor':RandomForestRegressor(),
       'SVR':SVR(kernel='rbf'),
       'KNeighborsRegressor':KNeighborsRegressor(n_neighbors=5)}
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score

def evall(true,pred):
    mse=mean_squared_error(true,pred)
    rmse=np.sqrt(mse)
    mae=mean_absolute_error(true,pred)
    r2=r2_score(true,pred)
    return mse,rmse,mae,r2
```

```python
def model_evall(x_train,x_test,y_train,y_test):
    for i in range(len(models.values())):
        
        model=list(models.values())[i]
        model.fit(x_train,y_train)
        train_pred=model.predict(x_train)
        test_pred=model.predict(x_test)
        print("For " + list(models.keys())[i]+" train prediction......................")
        mse,rmse,mae,r2=evall(y_train,train_pred)
        print('mse value is:',mse)
        print('rmse value is:',rmse)
        print('mae value is:',mae)
        print('r2 value is:',r2)
        print("  ")
        print("For "+list(models.keys())[i]+" test prediction......................")
        mse,rmse,mae,r2=evall(y_test,test_pred)
        print('mse value is:',mse)
        print('rmse value is:',rmse)
        print('mae value is:',mae)
        print('r2 value is:',r2)    
        print("  ")
        print("===========================================================")
        print("  ")
        print("  ")
```

```python
model_evall(x_train,x_test,y_train,y_test)
```
